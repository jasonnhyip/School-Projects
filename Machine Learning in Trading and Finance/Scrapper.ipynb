```python
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
# source = []
# description = []
# headline = []
# date = []

def get_page_items(url, items, idx):

  driver.get(url)
  driver.implicitly_wait(1000)
  content = driver.page_source
  soup = BeautifulSoup(content, 'lxml')

  source_tmp = [ x.get_text() for x in soup.find_all('span', attrs={'class': 'typography__StyledTypography-owin6q-0 hqfJpA capitalize'}) ][1:]
  description_tmp = [ x.get_text() for x in soup.find_all('p', attrs={'class': 'typography__StyledTypography-owin6q-0 lkUGSO'}) ][3:]
  headline_tmp = [ x.get_text() for x in soup.find_all('h6', attrs={'class': 'typography__StyledTypography-owin6q-0 cMKLWg'}) ]
  print(headline_tmp)
  # date = datetime.strptime(soup.find('time').text, "%B %d, %Y, %I:%M %p")
  date_author = [ x.get_text() for x in soup.find_all('h6', attrs={'class': 'typography__StyledTypography-owin6q-0 feyLb'}) ]
  # author = [ x.split(' ', 1)[1] for x in date_author[1::2] ]
  date_tmp = []
  for string in date_author:
      if "2021" in string:
        date_tmp.append(string)
  
  
  print(len(source_tmp), len(headline_tmp), len(description_tmp), len(date_tmp))
  if len(source_tmp) != len(headline_tmp) or len(source_tmp) != len(description_tmp) or len(source_tmp) != len(date_tmp):
    # print("source: ",  source_tmp)
    # print("headline: ", headline_tmp)
    # print("description: ", description_tmp)
    # print("date: ", date_tmp)
    # if idx == 17:
    if len(description_tmp) < len(source_tmp):
      for i in range(len(source_tmp) - len(description_tmp)):
        description_tmp.append("")
      print(f"page {idx+1}: missing description")
    if len(date_tmp) < len(source_tmp):
      for i in range(len(source_tmp) - len(date_tmp)):
        date_tmp.append("")
      print(f"page {idx+1}: missing date")

  for i in range(len(source_tmp)):
    news = {"Source": source_tmp[i], "Headline": headline_tmp[i], "Description": description_tmp[i], "Date": date_tmp[i]}
    items.append(news)
  # source.extend(source_tmp)
  # description.extend(description_tmp[3:])
  # headline.extend(headline_tmp)
  # date.extend(date_author[::2])

  # return list(zip(source, headline, description, date))
```

```python
import time

all_items = []

for i in range(195):
  
  pagination = f"&i={i}"
  url = f"https://www.coindesk.com/search?s=eth&sd=1609495409000&ed=1640945513000{pagination}&df=Custom"
  print(url)
  success = False;
  while not success:
    page_items = []
    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    get_page_items(url, page_items, i)
    driver.close()
    if i > 0 and page_items[0]['Headline'] == "Cardano vs. Ethereum: Can ADA Solve Ether’s Problems?":
      success = False
      # print("False: ", page_items)
    else:
      success = True
      all_items.extend(page_items)

  # except:
    # print('blocked')
  ##adding time sleep will decrease the chance of us being blocked, but increase the operation time
  # info.append(scraped)
  
  # time.sleep(1)
```
